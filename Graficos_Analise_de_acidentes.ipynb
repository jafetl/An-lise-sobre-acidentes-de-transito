```python
import pandas as pd
import numpy as np
import seaborn as sns
from collections import Counter
import collections
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/jafetl/An-lise-sobre-acidentes-de-transito/main/datatran2021.csv" , encoding = "ISO-8859-1", sep = ';')
```

```python
df = df.dropna()
sns.set(rc={'figure.figsize':(11.7,11)})
df['data_inversa']=  pd.to_datetime(df['data_inversa'])
```

```python
pessoas_veiculos = sns.scatterplot(data=df, x="pessoas", y="veiculos" ,hue="classificacao_acidente", palette="deep")
pessoas_veiculos.figure.set_size_inches(10,10)
```

```python
z = sns.scatterplot(data=df, x="pessoas", y="veiculos" ,hue="classificacao_acidente", palette="deep")
z.figure.set_size_inches(10,10)
```

## Correlao entre as variveis disponveis 

```python
sns.heatmap(df.corr(), annot = True, fmt=".1f")
```

## Quantidade de acidentes por estados

```python
sns.set(rc={'figure.figsize':(11.7,11)})
sns.countplot(data=df, x= 'uf')
```

```python
o = collections.UserDict(df['mortos'])
g = []

for key, value in o.items():
    if value != 0 : 
        g.append(key)
```

```python
mortos_df = pd.DataFrame(df.loc[g])
```

```python
alcool = df[df['causa_acidente'] == 'Ingesto de lcool pelo condutor']
```

```python
#Grfico de feridos em geral
sns.countplot(data=df, x= 'feridos')
```

```python
#Grafico de mortos por condio metereologica  
sns.histplot(data=mortos_df, x= 'condicao_metereologica',hue = 'condicao_metereologica')
```

# tipos de acidentes que causam mais mortes



```python
sns.countplot(data=mortos_df, y= 'tipo_acidente')
```

# Acidentes no geral


```python
sns.countplot(data=df, y= 'tipo_acidente')
```

## Quantidade de acidentes por Ingesto de lcool pelo condutor por horrio
